```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v = 40
t = np.linspace(0,10,10)
d = v * t
plt.plot(t,d,'y')
plt.xlabel('Time(seconds)')
plt.ylabel('Distance(metres)')
```

```python
plt.subplot(1,2,1)
plt.plot(t,d,'r')
plt.subplot(1,2,2)
plt.plot(d,t,'b')
```

```python
v = 40
t = np.linspace(0,10,10)
d = v * t
x = np.linspace(0,5,10)
y = x ** 2

plt.subplot(2,2,1)
plt.plot(t,d,'r')
plt.xlabel('Time(seconds)')
plt.ylabel('Distance(metres)')

plt.subplot(2,2,2)
plt.plot(d,t,'b')
plt.xlabel('Distance(metres)')
plt.ylabel('Time(seconds)')

plt.subplot(2,2,3)
plt.plot(x,y,'r')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2,2,4)
plt.plot(y,x,'b')
plt.xlabel('y')
plt.ylabel('x')
```

```python
#Object priented method matplotlib
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Object Oriented Programming')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes2.plot(x,y,'b')
```

```python

fig,axes = plt.subplots(2,2)
axes[0][0].plot(x,y,'r')
axes[0][1].plot(y,x,'b')
axes[1][0].plot(t,d,'b')
axes[1][1].plot(d,t,'b')
```

```python
grades = ["A","B","C","F"]
students = ["4", "8", "10", "6"]
plt.bar(grades,students,color="yellow", width = 0.8)
```

```python
grades_studs = {'A':4,'B':10,'C':10,'D':4,'F':15}
grades = list(grades_studs.keys())
no_studs = list(grades_studs.values())
plt.bar(grades,no_studs,color="yellow", width = 0.5)
plt.yticks(np.arange(0,20,1))
plt.grid(color='black',linewidth=0.5,alpha=0.2)
```

```python
#Bar plot object oriented method
fig,axes = plt.subplots()
axes.bar(grades,no_studs,color='maroon',width=0.5)
axes.set_yticks(np.arange(0,16,1))
```
